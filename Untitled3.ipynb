```python
df = pd.read_csv("//content/drive/MyDrive/MCD/mcdonalds.csv")
print(df)
```

```python
MD_x = df.iloc[:, :11]
MD_x = (MD_x == "Yes").astype(int)
col_means = MD_x.mean()
rounded_means = np.round(col_means, 2)
print(rounded_means)
```

```python
pca = PCA()
MD_pca = pca.fit(MD_x)
```

```python
explained_variance = pca.explained_variance_
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)
```

```python
summary_df = pd.DataFrame( # Creating an instance of DataFrame
{
    'Standard Deviation': np.sqrt(explained_variance),
    'Proportion of Variance': explained_variance_ratio,
    'Cumulative Proportion': cumulative_variance_ratio
})
summary_df = summary_df.round(4)
print(summary_df)
```

```python
standard_deviations = np.sqrt(MD_pca.explained_variance_)
```

```python
print(np.round(standard_deviations, 1))
```

```python
rotation_matrix = MD_pca.components_.T
print(rotation_matrix.shape)
print(rotation_matrix)
```

```python
features = MD_x.columns  # Assuming MD_x is a DataFrame; if not, define the feature names
pcs = [f"PC{i+1}" for i in range(rotation_matrix.shape[1])]
rotation_df = pd.DataFrame(rotation_matrix, index=features, columns=pcs)
```

```python
rotation_df = rotation_df.round(3)
```

```python
print(rotation_df)
```

```python
scaler = StandardScaler()
MD_x_scaled = scaler.fit_transform(MD_x)
```

```python
pca = PCA()
MD_pca = pca.fit_transform(MD_x_scaled)
```

```python
pca_df = pd.DataFrame(data=MD_pca, columns=[f"PC{i+1}" for i in range(MD_pca.shape[1])])
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(pca_df['PC1'], pca_df['PC2'], color='grey', alpha=0.5)  # Plot first two PCs
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Results')
plt.grid(True)
```

```python
for i in range(len(MD_x.columns)):
    plt.arrow(0, 0, pca.components_[0, i]*max(pca_df['PC1']), pca.components_[1, i]*max(pca_df['PC2']),
             head_width=0.05, head_length=0.1, fc='r', ec='r')
    plt.text(pca.components_[0, i]*max(pca_df['PC1']), pca.components_[1, i]*max(pca_df['PC2']),
             MD_x.columns[i], color='r')

plt.show()
```
